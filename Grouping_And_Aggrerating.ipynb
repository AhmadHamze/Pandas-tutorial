```python
import pandas as pd
df = pd.read_csv('data/survey_results_public.csv')
```

```python
country_grp = df.groupby(['Country'])
```

```python
country_grp.get_group('United States')
```

```python
country_grp['SocialMedia'].value_counts(normalize=True).head(50) #this is a series with two indexes
```

```python
country_grp['ConvertedComp'].agg(['mean', 'median'])
```

```python
# How many people use Python in each country, without 'sum' you will get a very different result
nbr_of_python_users = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
nbr_of_python_users
```

```python
nbr_of_people_country = df['Country'].value_counts()
nbr_of_people_country
```

```python
python_df = pd.concat([nbr_of_people_country, nbr_of_python_users], axis='columns')
python_df.rename(columns={'Country': 'NbrOfParticipants', 'LanguageWorkedWith': 'NbrPythonUsers'}, inplace=True)
python_df['PctPythonUsers'] = (nbr_of_python_users / nbr_of_people_country) * 100
python_df
```
