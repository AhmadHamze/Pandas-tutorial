```python
import pandas as pd
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent')
df_schema = pd.read_csv('data/survey_results_schema.csv', index_col='Column')
df.head()
```

```python
#df['Hobbyist'] = df['Hobbyist'].astype(bool) # Both 'yes' and 'no' will become True!
df['Hobbyist'].replace({'Yes': True, 'No': False}, inplace=True)
```

```python
df_temp = df[['Country', 'Hobbyist', 'Employment']]
df_temp
```

```python
df_temp.dropna()
```

```python
df_temp['Employment'].unique()
```

```python
# Any row containing at least one NaN will be removed
df_temp.dropna()
```

```python
# Any row containing at least one NaN IN THE EMPLOYMENT COLUMN is removed, other columns are ignored
df_temp.dropna(axis='index', how='any', subset=['Employment'])
```

```python
# The rows that will be removed are the ones having NaN in both columns Country and Employment
df_temp.dropna(axis='index', how='all', subset=['Country', 'Employment'])
```
