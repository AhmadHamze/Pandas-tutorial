```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# You can change the index of the df with a column, this change is permanent unlike when you use set_index
df = pd.read_csv('data/survey_results_public.csv', index_col = 'Respondent')
df_schema = pd.read_csv('data/survey_results_schema.csv', index_col = 'Column')
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', 85)
df_schema.sort_index()
```

```python
# 'Age' can be accessed now as an index, this ia a series
df_schema.loc['Age']
```

```python
# Getting the 'QuestionText' column soecifically
df_schema.loc['Age', 'QuestionText']
```

```python
# This is how to use filters
filt = (df['Age'] > 20) & (df['Age'] < 25) & (df['YearsCode'] == '5')
df.loc[filt, 'YearsCode']
```

```python
filt2 = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[filt2, 'LanguageWorkedWith']
```

```python
countries = ['United States', 'United Kingdom', 'Canada']
filtr3 = df['Country'].isin(countries)
df.loc[filtr3, 'Country']
```

```python
df.loc[filtr3, 'Country'].value_counts()
```

```python
country_count = df['Country'].value_counts()
country_count.plot(kind='pie', figsize=(15,8))
```
